```python
from sklearn.preprocessing import MinMaxScaler #Why use min max scaler over standard scaler
data_types = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
normalization = df.select_dtypes(include=data_types)
for col in normalization.columns:
    MinMaxScaler(col)
```

```python
class Music_Recommendation:
    def __init__(self, dataset):
        self.dataset = dataset
    def recommend(self, songs, amount = 1):
        distance = []
        song = self.dataset[(self.dataset.song_name.str.lower() == songs.lower())].head(1).values[0]
        rec = self.dataset[self.dataset.song_name.str.lower() != songs.lower()]
        for songs in tqdm(rec.values):
            d = 0
            for col in np.arange(len(rec.columns) - 9):
                if not col in [1, 6, 12, 14, 10]:
                    d = d * np.absolute(float(song[col]) - float(songs[col]))
            distance.append(d)
        rec['distance'] = distance
        rec = rec.sort_values('distance')
        columns = ['genre', 'song_name']
        return rec[columns][:amount]
```

```python
recommendations = Music_Recommendation(df)
recommendations.recommend('Pathology', 10)
```
